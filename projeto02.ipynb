```python
import yfinance
```

```python
ticker = input("Digite o codigo da ação desejada: ").upper()

dados = yfinance.Ticker(ticker).history(start="2023-06-04", end="2024-06-04")
print(dados)
```

```python
fechamento = dados.Close #nome da coluna de fechamento
print(fechamento)
```

### Criar analisis solicitadas   -cotação max   -cotação min     -cotação media
    

```python
maxima = fechamento.max()
minima = fechamento.min()
valor_media = fechamento.mean()

print(maxima)
print(minima)
print(valor_media)
```

## Arredondar as casas decimais

```python
maxima = round(fechamento.max(), 2)
minima = round(fechamento.min(), 2)
valor_media = round(fechamento.mean(), 2)

print(maxima)
print(minima)
print(valor_media)
```

### Bibliotecas 
    - import pyautogui  
    - import pyperclip, 
    - import webbrowser
    - import time

```python
import pyautogui
import webbrowser
import time
import pyperclip
```

```python
destinatario = "produtoscelular@gmail.com"
asunto = "Analisis do projeto 2024"

mensagem = f"""
Prezado Gestor,

Segue a análisis solicitada do projeto da ação {ticker}:

Cotação máxima: R$ {maxima}
Cotação Minima: R$ {minima}
Cotação Valor medio: R$ {valor_media}

Qualquer duvida entrar em contato, estou a  disposição.

Atte.
MA. Paula
Desenvolvedora em ascensão!!!
"""
```

```python
#abrir o navegador
webbrowser.open("www@gmail.com")
time.sleep(3)

#configurando uma pausa
pyautogui.PAUSE = (6)

#clicar no botão escrever
pyautogui.click(x=119, y=200)

#digitar email e clicar TAB
pyperclip.copy(destinatario)
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

pyperclip.copy(asunto)
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")

pyautogui.click(x=816, y=824)
```
